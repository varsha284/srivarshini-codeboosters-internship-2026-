```python
1. Explain the 5 Vs of Big Data using Flipkart as an example
1. Volume

Big Data contains huge amounts of data.

Example in Flipkart:

Millions of users browse products daily
Product details, orders, reviews, payments, and delivery records generate terabytes of data
2. Velocity

Data is generated and processed very fast.

Example:

Real-time order updates
Live inventory tracking
Instant recommendations while shopping
3. Variety

Data comes in different formats.

Example:

Structured data → customer tables, sales records
Semi-structured → JSON logs
Unstructured → product images, reviews, videos
4. Veracity

Data quality and accuracy matter.

Example:

Fake reviews
Incorrect addresses
Duplicate customer accounts
Flipkart must clean and validate data before analysis.
5. Value

Useful insights are extracted from data.

Example:

Personalized product recommendations
Demand prediction
Fraud detection
Better delivery

2. Difference between Transformation and Action in PySpark
Transformation	Action
Creates a new DataFrame/RDD	Produces final output
Lazy execution	Triggers execution
Does not execute immediately	Executes immediately
Used for data processing	Used for result/output
Examples
Transformations
df.filter(df.age > 18)
df.select("name")
df.groupBy("city")
Actions
df.show()
df.collect()
df.count()
Important Point

PySpark follows lazy evaluation.
Transformations are executed only when an Action is called.

3. PySpark code to find the top 5 cities by total revenue
from pyspark.sql.functions import sum

# Read dataset
df = spark.read.csv("sales_data.csv", header=True, inferSchema=True)

# Group by city and calculate total revenue
top_cities = df.groupBy("city") \
               .agg(sum("revenue").alias("total_revenue")) \
               .orderBy("total_revenue", ascending=False) \
               .limit(5)

# Show result
top_cities.show()

4. Why is Parquet preferred over CSV for Big Data?
1. Columnar Storage

f(x)=Read Only Required Columns

Parquet stores data column-wise, so Spark reads only required columns.
CSV reads the entire file.

Result:

Faster queries
Better performance
2. Compression

Parquet supports efficient compression.

Result:

Smaller storage size
Reduced disk usage
3. Better Schema Support

Parquet stores schema information automatically.

Result:

Correct data types
Easier processing
No repeated schema inference

5. Explain the Medallion Architecture — what lives in each layer?

Medallion Architecture organizes data into layers for better quality and processing.

Bronze Layer (Raw Data)
Original raw data
Data from APIs, logs, CSV files, databases
Minimal processing

Example:

Raw sales transactions
Silver Layer (Cleaned Data)
Cleaned and transformed data
Removed duplicates
Fixed null values
Standardized formats

Example:

Validated customer and sales data
Gold Layer (Business-Level Data)
Aggregated and analytics-ready data
Used for dashboards and reporting

Example:

Top-selling products
Monthly revenue reports
Flow
Bronze → Silver → Gold

6. Difference between a Data Lake and a Data Warehouse
Data Lake	Data Warehouse
Stores raw data	Stores processed data
Supports structured, semi-structured, unstructured data	Mostly structured data
Flexible schema	Fixed schema
Used for Big Data and ML	Used for reporting and BI
Cheap storage	More expensive
Example: Hadoop, S3	Example: Snowflake, Redshift
Simple Understanding
Data Lake
Store everything first, process later
Data Warehouse
Process first, then store clean structured data
```
