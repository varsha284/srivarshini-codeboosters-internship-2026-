```python
import os
import streamlit as st
from huggingface_hub import InferenceClient

# Page configuration
st.set_page_config(
    page_title="Book Suggestion Assistant",
    page_icon="📚"
)

# Title
st.title("📚 Book Suggestion Assistant")
st.write("Get book recommendations using Llama 3.")

# Hugging Face API Token
HF_TOKEN = os.getenv("HF_TOKEN")

if not HF_TOKEN:
    st.error("Please set your Hugging Face token.")
    st.stop()

# Model
MODEL_ID = "meta-llama/Meta-Llama-3-8B-Instruct"

client = InferenceClient(
    api_key=HF_TOKEN,
    model=MODEL_ID
)

def extract_text(resp):
    try:
        return resp.choices[0].message.content
    except Exception:
        try:
            return resp["choices"][0]["message"]["content"]
        except Exception:
            return str(resp)

# User Input
query = st.text_area(
    "Enter your book-related question:",
    "List 10 books written by Indian author Chetan Bhagat."
)

if st.button("Get Recommendations"):
    with st.spinner("Generating response..."):
        messages = [
            {
                "role": "system",
                "content": "You are the best book suggesting assistant."
            },
            {
                "role": "user",
                "content": query
            }
        ]

        try:
            resp = client.chat_completion(
                messages=messages,
                max_tokens=400
            )

            answer = extract_text(resp)

            st.subheader("Response")
            st.write(answer)

        except Exception as e:
            st.error(f"Error: {e}")

# Save the Streamlit code to a file
with open('app.py', 'w') as f:
    f.write('''
import os
import streamlit as st
from huggingface_hub import InferenceClient

# Page configuration
st.set_page_config(
    page_title="Book Suggestion Assistant",
    page_icon="📚"
)

# Title
st.title("📚 Book Suggestion Assistant")
st.write("Get book recommendations using Llama 3.")

# Hugging Face API Token
HF_TOKEN = os.getenv("HF_TOKEN")

if not HF_TOKEN:
    st.error("Please set your Hugging Face token.")
    st.stop()

# Model
MODEL_ID = "meta-llama/Meta-Llama-3-8B-Instruct"

client = InferenceClient(
    api_key=HF_TOKEN,
    model=MODEL_ID
)

def extract_text(resp):
    try:
        return resp.choices[0].message.content
    except Exception:
        try:
            return resp["choices"][0]["message"]["content"]
        except Exception:
            return str(resp)

# User Input
query = st.text_area(
    "Enter your book-related question:",
    "List 10 books written by Indian author Chetan Bhagat."
)

if st.button("Get Recommendations"):
    with st.spinner("Generating response..."):
        messages = [
            {
                "role": "system",
                "content": "You are the best book suggesting assistant."
            },
            {
                "role": "user",
                "content": query
            }
        ]

        try:
            resp = client.chat_completion(
                messages=messages,
                max_tokens=400
            )

            answer = extract_text(resp)

            st.subheader("Response")
            st.write(answer)

        except Exception as e:
            st.error(f"Error: {e}")
''')
```
